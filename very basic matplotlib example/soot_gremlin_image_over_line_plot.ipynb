```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Example line plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(x, y, label='Soot Gremlin Performance')

# Load the image
img = mpimg.imread('susuwatari_soot_gremlin.jpg')  

# Create an OffsetImage
imagebox = OffsetImage(img, zoom=0.4)  # You can control the size of the image with 'zoom'

# Create an AnnotationBbox - it will place the image at the position (2, 4) on the plot
ab = AnnotationBbox(imagebox, (4, 16), frameon=False, boxcoords="data", pad=0)

# Add the AnnotationBbox to the ax
ax.add_artist(ab)

# Labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Coal Revenue')
ax.legend()

# Show the plot
plt.show()
```
